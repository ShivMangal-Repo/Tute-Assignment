Load Data from SQL Database (SQLite)

```python
conn = sqlite3.connect("sample.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER,
    name TEXT,
    department TEXT,
    salary INTEGER
)
""")
```

```python
cursor.executemany("""
INSERT INTO employees VALUES (?, ?, ?, ?)
""", [
    (1, "Amit", "IT", 60000),
    (2, "Neha", "HR", 45000),
    (3, "Rahul", "Finance", 70000),
    (4, "Pooja", "IT", 65000),
    (5, "Ankit", "Sales", 50000)
])
```

```python
conn.commit()
```

```python
sql_df = pd.read_sql("SELECT * FROM employees", conn)
sql_df
```
