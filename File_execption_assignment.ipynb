##Task1:Division Utility

```python

try:
    num = float(input("Enter numerator: "))
    den = float(input("Enter denominator: "))
    result = num / den
except ValueError:
    print("Error: Please enter numeric values.")
except ZeroDivisionError:
    print("Error: Denominator cannot be zero.")
else:
    print("Result:", result)
finally:
    print("Operation Complete")
```

##Task2: Bill Calculator

```python

prices = [120, 350, 'abc', 500, -200, 800]
total = 0.0

for item in prices:
    try:

        price = float(item)

        if price < 0:
            raise ValueError("Negative price not allowed")

        total += price

    except TypeError:
        print(f"Skipped (TypeError): {item}")
    except ValueError as e:
        # Handles both conversion errors and our custom negative-price error
        print(f"Skipped ({e.__class__.__name__}): {item} -> {e}")

print("Running total:", total)
```

##Task3: Custom Exception

```python

def check_age(age):
    if not (1 <= age <= 120):
        raise ValueError("Age must be between 1 and 120")
    return True

age_input = input("Enter your age: ")

try:
    age_val = int(age_input)  # may raise ValueError
    check_age(age_val)        # may raise ValueError (custom)
    print("Age accepted.")
except ValueError as e:
    print("Error:", e)
```

##Task4: File Reader with Exception Handling

```python

filename = input("Enter filename to read: ")

try:
    f = open(filename, "r")  # may raise FileNotFoundError or PermissionError
    try:
        # Read first 3 lines (print fewer if file has < 3 lines)
        for i in range(3):
            line = f.readline()
            if not line:
                break
            print(line.rstrip("\n"))
    finally:
        # Ensure the file is closed if it was opened successfully
        f.close()

except FileNotFoundError:
    print("Error: File not found.")
except PermissionError:
    print("Error: Permission denied.")
finally:
    print("File operation attempted.")
```

##Task5: Mini Program

```python

cart = []
print("Enter item prices. Type 'q' to finish.")

while True:
    raw = input("Price: ").strip()
    if raw.lower() == 'q':
        break

    try:
        price = float(raw)  # may raise ValueError

        if price < 0:
            # Custom exception for negative inputs
            raise ValueError("Negative price not allowed")

        cart.append(price)

    except ValueError as e:
        print("Invalid input:", e)

# Summary
total_items = len(cart)
total_bill = sum(cart)

print("Total items:", total_items)
print("Total bill:", total_bill)
```
