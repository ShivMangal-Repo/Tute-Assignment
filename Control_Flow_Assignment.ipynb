#Task 1: Discount Rules (if / elif / else)

```python
# Read input
order_input = input("Enter order amount: ")

# Check if input is numeric
if order_input.isdigit():
    order_amount = int(order_input)

    # Apply discount rules
    if order_amount >= 2000:
        discount_percent = 15
    elif order_amount >= 1500:
        discount_percent = 10
    elif order_amount >= 1000:
        discount_percent = 7
    else:
        discount_percent = 0

    discount_amount = order_amount * discount_percent / 100
    subtotal = order_amount - discount_amount

    print("Order Amount:", order_amount)
    print("Discount Applied:", discount_percent, "%")
    print("Subtotal after discount:", subtotal)

    # Extra (optional): Add tax
    tax = subtotal * 5 / 100
    final_total = subtotal + tax

    print("Tax (5%):", tax)
    print("Final Amount:", final_total)

else:
    print("Error: Please enter a valid numeric amount")
```

#Task 2: Process Multiple Orders (for loop)

```python
orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discounted_orders_count = 0

print("Order | Discount% | Final Amount")
print("---------------------------------")

for order_amount in orders:

    # Apply discount rules
    if order_amount >= 2000:
        discount_percent = 15
    elif order_amount >= 1500:
        discount_percent = 10
    elif order_amount >= 1000:
        discount_percent = 7
    else:
        discount_percent = 0

    if discount_percent > 0:
        discounted_orders_count += 1

    discount_amount = order_amount * discount_percent / 100
    final_amount = order_amount - discount_amount

    total_revenue += final_amount

    print(order_amount, " | ", discount_percent, "% | ", final_amount)

print("---------------------------------")
print("Total revenue after discounts:", total_revenue)

# Extra (optional)
print("Orders that received discount:", discounted_orders_count)
```

#Task 3: User Menu (while loop + break / continue)

```python
orders = []

while True:
    print("\nMenu")
    print("1 - Add order amount")
    print("2 - Show orders and totals after discount")
    print("q - Quit")

    choice = input("Enter your choice: ")

    if choice == "q":
        print("Exiting program.")
        break

    elif choice == "1":
        amount_input = input("Enter order amount: ")

        if amount_input.isdigit():
            orders.append(int(amount_input))
            print("Order added.")
        else:
            print("Invalid amount.")
            continue

    elif choice == "2":
        if len(orders) == 0:
            print("No orders available.")
            continue

        total = 0
        print("\nOrder | Discount% | Final Amount")
        print("--------------------------------")

        for order_amount in orders:
            if order_amount >= 2000:
                discount = 15
            elif order_amount >= 1500:
                discount = 10
            elif order_amount >= 1000:
                discount = 7
            else:
                discount = 0

            final_amount = order_amount - (order_amount * discount / 100)
            total += final_amount

            print(order_amount, "|", discount, "% |", final_amount)

        print("--------------------------------")
        print("Total after discounts:", total)

    else:
        print("Invalid menu option.")
        continue
```

#Task ***: Loop Control with Conditions (break & continue)

```python
daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for sale in daily:

    if sale == -1:
        print("Corrupted data found. Stopping processing.")
        break

    if sale == 0:
        print("No sales today. Skipping.")
        continue

    total_sales += sale
    print("Running total:", total_sales)

print("Final total sales:", total_sales)
```
