```python
# Parallel category list
categories = [
    "Electronics",
    "Electronics",
    "Electronics",
    "Electronics",
    "Stationery",
    "Stationery"
]

# Create set from categories
categories_set = set(categories)

print("Unique categories:", categories_set)

# Add a new category
categories_set.add("Accessories")

print("Updated categories set:", categories_set)
```

```python
# Check if a category exists in the set
check_category = "Electronics"

print(check_category in categories_set)
```

Extra (Optional): Total number of unique categories

```python
# Get total number of unique categories
total_categories = len(categories_set)

print("Total unique categories:", total_categories)
```

#Task3

```python
#Create the price dictionary (6 products)
price_dict = {
    "Laptop": 55000,
    "Headphones": 2000,
    "Keyboard": 1500,
    "Mouse": 800,
    "Notebook": 150,
    "Pen": 20
}
```

```python
#Add a new product
price_dict["Charger"] = 1200

print("After adding new product:", price_dict)
```

```python
#Update price of an existing product
price_dict["Mouse"] = 750

print("After updating price:", price_dict)
```

```python
#Remove a product
product_to_remove = "Tablet"
if product_to_remove in price_dict:
    del price_dict[product_to_remove]
    print(product_to_remove, "removed")
else:
    print(product_to_remove, "does not exist")
```

```python
#Print average price of all products
total_price = 0

for price in price_dict.values():
    total_price += price

average_price = total_price / len(price_dict)

print("Average product price:", average_price)
```

```python
#Extra (Optional): Product with maximum and minimum price
max_product = None
min_product = None

for product in price_dict:
    if max_product is None or price_dict[product] > price_dict[max_product]:
        print("Highest priced product:", product, price_dict[product])
        max_product = product
        print("Highest priced product:", max_product, price_dict[max_product])
    if min_product is None or price_dict[product] < price_dict[min_product]:
        min_product = product

print("Highest priced product:", max_product, price_dict[max_product])
print("Lowest priced product:", min_product, price_dict[min_product])
```
